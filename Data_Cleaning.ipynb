What are the potential consequences of analyzing unclean or messy data?

Explain the common steps involved in cleaning and organizing data.

Answere- Data cleaning is the process of identifying, correcting, or removing inaccurate, incomplete, inconsistent, duplicated, or irrelevant data before analysis.
Data analysis is only as good as the data behind it. Cleaning ensures that:

Results are accurate and reliable.
Insights are meaningful.
Models perform better.
Decisions are trustworthy.
Time and resources are saved.

Analyzing dirty data can lead to serious problems, such as:

Incorrect conclusions.
Biased analysis.
Poor predictive models.
Loss of credibility.

Ques 5- Apply Data Validation to restrict Quantity values to only whole numbers between 1 and 10.
Configure an input message that appears when a user selects a cell in the "Quantity" column explaining:
"Please enter a whole number between 1 and 10."
Set up an error alert message that triggers if the user enters a number less than 1 or greater than 10,
showing:
"Invalid input! The quantity must be a whole number between 1 and 10."
Write a step-by-step approach for this question

Answere- Step 1- Identify the Quantity column
Locate the Quantity column in your dataset.
Select all the cells under the Quantity column where users will enter values.
Step 2- Open Data Validation
Go to the Data tab on the Excel ribbon.
Click Data Validation.
In the dropdown, choose Data Validation…
Step 3- Set validation criteria
In the Settings tab:
Allow: Select Whole number
Data: Select Between
Minimum: Enter 1
Maximum: Enter 10
Step 4- Configure the Input Message
Click the Input Message tab.
Check Show input message when cell is selected.
Enter:
Title: Quantity Entry
Input Message:Please enter a whole number between 1 and 10.
Step 5- Configure the Error Alert
Click the Error Alert tab.
Check Show error alert after invalid data is entered.
Set:
Style: Stop
Title: Invalid Input
Error Message:Invalid input! The quantity must be a whole number between 1 and 10.
